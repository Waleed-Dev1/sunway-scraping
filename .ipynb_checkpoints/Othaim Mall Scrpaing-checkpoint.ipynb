```python
def process_google_api_fetching(store, iteration):
    if iteration % 100 == 0:
        print(f"Processing records Number: {iteration}")
    
    # Create an empty DataFrame for this specific store
    detail = pd.DataFrame(index=[0])  # Single row DataFrame
    
    # First, add all the original store data
    for column in store.index:
        detail[column] = store[column]
    
    try:
        store_geocode = gmaps.geocode(f"{store['Brand']} othaim mall {store['Mall']}, Saudi Arabia")
        place_id = store_geocode[0]['place_id']
        
        details_url = f"https://places.googleapis.com/v1/places/{place_id}"
        params = {
            "fields": "displayName,rating,reviews,international_phone_number,current_opening_hours,nationalPhoneNumber,formattedAddress,types,photos",
            "key": API_KEY
        }
        response = requests.get(details_url, params=params)
        data = response.json()
        
        # Get types list from API response
        types_list = data.get('types', [])
        
        # Check if unwanted categories are in the types list
        unwanted_categories = ["sublocality_level_1", "sublocality", "political", "locality", "route"]

        # For unwanted categories, return store data with "No data found" for Google data
        if any(category in types_list for category in unwanted_categories):
            for column in ['google_name', 'google_category', 'google_location', 'google_reviews', 
                        'google_rating', 'google_nationalPhoneNumber', 'google_opening_hours',
                        'google_internationalPhoneNumber', 'google_place_id']:
                detail[column] = 'No data found'
            
            # Add a flag to indicate unwanted category
            # detail["unwanted_category_detected"] = True
            # detail["detected_types"] = ", ".join(types_list)
            return detail
        
        # Continue with normal processing for valid categories
        detail["google_name"] = data.get('displayName', {}).get('text', 'N/A')
        detail["google_category"] = ", ".join(types_list)
        
        detail["google_location"] = data.get('formattedAddress', 'N/A')
        
        # Process reviews
        reviews_list = []
        for rev in data.get('reviews', []):
            review_text = rev.get("originalText", {}).get("text", "No Review")
            review_rating = rev.get("rating", "N/A")
            review_time = rev.get("relativePublishTimeDescription", "N/A")
            reviews_list.append(f"review_rating:{review_rating}, text:{review_text}")
        
        detail["google_reviews"] = " | ".join(reviews_list) if reviews_list else "No Reviews"
        
        detail["google_rating"] = data.get('rating', 'N/A')
        detail["google_nationalPhoneNumber"] = data.get('nationalPhoneNumber', 'N/A')
        
        opening_hours = data.get('currentOpeningHours', {}).get("weekdayDescriptions", ["N/A"])
        detail["google_opening_hours"] = ", ".join(opening_hours)
        
        detail["google_internationalPhoneNumber"] = data.get('internationalPhoneNumber', 'N/A')
        detail["google_place_id"] = place_id
        
        return detail
        
    except Exception as e:
        logger.error(f"Error processing store {store['Brand']} in {store['Mall']}: {e}")
        # Return DataFrame with original store data but empty Google data
        for column in ['google_name', 'google_category', 'google_location', 'google_reviews', 
                       'google_rating', 'google_nationalPhoneNumber', 'google_opening_hours',
                       'google_internationalPhoneNumber', 'google_place_id', 'unwanted_category_detected']:
            detail[column] = 'Error'
        detail["error_message"] = str(e)
        return detail
```

```python
import logging

# Configure your logger
logger = logging.getLogger(__name__)

# Initialize an empty list to store futures
futures = []

# Initialize an empty DataFrame to store all results
all_results = pd.DataFrame()

# Execute with ThreadPoolExecutor
with ThreadPoolExecutor(max_workers=4) as executor:
    futures_to_stores = {}
    for row_index, (_, store) in enumerate(shops.iterrows()):
        future = executor.submit(
            process_google_api_fetching, store, row_index
        )
        futures_to_stores[future] = store
    
    for future in as_completed(futures_to_stores):
        try:
            result_df = future.result()
            all_results = pd.concat([all_results, result_df], ignore_index=True)
        except Exception as e:
            store = futures_to_stores[future]
            logger.error(f"Error in future execution for store {store['Brand']} in {store['Mall']}: {e}", exc_info=True)

# Now all_results contains all the data
print(f"Collected data for {len(all_results)} stores")
```
